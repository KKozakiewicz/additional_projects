```python
import requests
import pandas as pd
from collections import Counter
import numpy as np
from sklearn import preprocessing
import seaborn as sns

from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
url = 'https://justjoin.it/api/offers'
```

```python
response = requests.get(url)
content = response.json()
```

```python
content[0]
```

```python
df = pd.DataFrame.from_records(content)
```

```python
df.sample(10)
```

```python
df.to_csv('raw_data.csv', index = False)
```

```python
df_employment_types = df['employment_types'].apply(pd.Series)
```

```python
df_employment_types.head()
```

```python
df_employment_type_1 = df_employment_types[0].apply(pd.Series)
```

```python
df_employment_type_1.sample(5)
```

```python
df_employment_type_1.columns
```

```python
df_employment_type_1= pd.concat([df_employment_type_1.drop(['salary'], axis=1), df_employment_type_1['salary'].apply(pd.Series)], axis=1)
```

```python
df_employment_type_1
```

```python
df_employment_type_2 = df_employment_types[1].apply(pd.Series)
```

```python
df_employment_type_2 = pd.concat([df_employment_type_2.drop(['salary'], axis=1), df_employment_type_2['salary'].apply(pd.Series)], axis=1)
```

```python
df_employment_type_2.drop(columns=0, inplace = True)
df_employment_type_2
```

```python
df_no_employment = df.drop(columns = 'employment_types')
```

```python
df_employment_1 = pd.concat([df_no_employment, df_employment_type_1], axis = 1)
df_employment_1
```

```python
df_employment_2 = pd.concat([df_no_employment, df_employment_type_2], axis = 1)
df_employment_2
```

```python
df_employment = pd.concat([df_employment_1, df_employment_2])
df_employment
```

```python
df_employment = df_employment.dropna(subset=['type'])
```

```python
skills_unpack = pd.DataFrame()
```

```python
skills_unpack = df_employment['skills'].apply(pd.Series)
```

```python
skills_unpack
```

```python
skills_unpack_0 = skills_unpack[0].apply(pd.Series)
skills_unpack_1 = skills_unpack[1].apply(pd.Series)
skills_unpack_2 = skills_unpack[2].apply(pd.Series)
```

```python
skills_unpack_2
```

```python
skills_unpack_1=skills_unpack_1.drop(columns=0)
```

```python
skills_unpack_2=skills_unpack_2.drop(columns=0)
```

```python
skills_unpack_0.rename(columns={'name': 'skill_1', 'level': 'skill_1_level'}, inplace=True)
skills_unpack_1.rename(columns={'name': 'skill_2', 'level': 'skill_2_level'}, inplace=True)
skills_unpack_2.rename(columns={'name': 'skill_3', 'level': 'skill_3_level'}, inplace=True)
```

```python
skills = pd.concat([skills_unpack_0, skills_unpack_1, skills_unpack_2], axis=1)
skills
```

```python
df_clean = pd.concat([df_employment, skills], axis = 1)
```

```python
df_clean.drop(columns = 'skills', inplace = True)
```

```python
df_clean.to_csv('clean_data.csv', index=False)
```
